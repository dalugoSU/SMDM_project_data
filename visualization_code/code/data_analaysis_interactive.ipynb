```python
data_all_gen = pd.concat([data_2018, data_2019, data_2020, data_2021, data_2022])
data_all_corp = pd.concat([data_facebook, data_amazon, data_apple, data_microsoft])

mod_data_all_corp = data_all_corp.drop("LanguageURL", axis=1)

data_all_merged = pd.concat([data_all_gen, mod_data_all_corp])
```

```python
# View Language statistics for x attribute for a specific year or over the course of 2018-2022
# From a collection of repos from general public or big corp
quantity_options = ["Language", "5 Languages", "10 Languages"]
from_where = ["Gen Repos Only", "Big Corp Only", "Both"]
see_as = ["Dataframe", "Diagram"]
data_options = ["Stars", 
                "Forks", 
                "WatchCount", 
                "NetworkCount", 
                "IssueCount", 
                "PRCount", 
                "ProjectsCount",
                "BranchCount",
                "ContributorCount"]

def show_diagram(data, attribute, head_count):
    # Copy frame and select attribute to explore
    language_data = data
    row_amount = head_count # Number of rows to plot

    # set figure size
    fig, ax = plt.subplots(figsize=(5, 7))

    # Select color for edge
    plt.rcParams['axes.edgecolor'] = '#fc4293'

    # Remove edge on right and left sides of graph
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    # ax.tick_params(axis='both', which='both', labelsize=10, bottom=True, left=False)

    # remove grid and set background color
    ax.grid(False)
    ax.set_facecolor('white')

    # plot the data from language head
    # Select head() for top amount to showcase
    # Data is already sorted by stars so most starred will be at the top
    # The more rows to showcase the height parameter in figsize should be adjusted as well
    # i.e 10 for head(50)
    sns.barplot(data=language_data.head(row_amount), x=attribute, y='Language');

    # X/Y axis titles and style
    ax.set_xlabel(f'{attribute}', fontsize=13, color='#8f0d18')
    ax.set_ylabel('Language', fontsize=13, color='#8f0d18')
    fig.suptitle(f'Average {attribute}; Top {row_amount} Languages in Big Corp', fontsize=13, color='#8f0d18');



@interact(From=from_where,
          Top=quantity_options, 
          Year=sorted(list(data_all_merged['Published'].unique()), reverse=True) + ['All'], 
          Most=data_options,
          View=see_as)
def filter_dataframe(From, Top, Year, Most, View):
    
    # Input variables
    i_data = None
    head_count = None
    
    if From == "Gen Repos Only":
        i_data = data_all_gen
    elif From == "Big Corp Only":
        i_data = data_all_corp
    elif From == "Both":
        i_data = data_all_merged
    
    if Top == "5 Languages":
        head_count = 5
    elif Top == "10 Languages":
        head_count = 10
    else:
        head_count = 1

        
    if Year != 'All':
        i_data = i_data[i_data['Published'] == Year] # filter rows for specific year
    
    
    # group by language and sum x filter
    grouped = i_data.groupby('Language')[Most].sum().reset_index()
    
    
    # sort by x filter in descending order
    sorted_grouped = grouped.sort_values(by=Most, ascending=False).reset_index(drop=True)[['Language', Most]]
    sorted_grouped.index += 1 # Index starts from 1
    
    
    if Top == "5 Languages":
        head_count = 5
    elif Top == "10 Languages":
        head_count = 10
    else:
        head_count = 1

    if View == "Dataframe":
        return sorted_grouped.head(head_count)
    elif View == "Diagram":
        show_diagram(sorted_grouped, Most, head_count)
```
